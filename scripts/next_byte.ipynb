```python
import matplotlib.pyplot as plt
import numpy as np
from spa.core import spa
from spa.properties import ThresholdProperty
import pandas as pd
```

```python
data = pd.read_csv('../data/counts.csv')
correct_list = data['is_correct'].values
# Split into 16 lists
split_list = np.split(correct_list, len(correct_list)/16)
split_list
```

```python
probabilities = []
for byte_list in split_list:
    num_samples = 0
    num_correct_after = 0
    for i, is_correct in enumerate(byte_list):
        if not is_correct or i == 15:
            continue
        num_samples += 1
        if byte_list[i + 1]:
            num_correct_after += 1
        probabilities.append(num_correct_after / num_samples)

print(probabilities)
```
